```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_genot_100_train = pd.read_csv("pancreas_unbalanced_100_genot_train.csv")
df_genot_100_train["model"] = "UGENOT-K 1.0"
df_genot_100_train["data"] = "train"
df_genot_999_train = pd.read_csv("pancreas_unbalanced_999_genot_train.csv")
df_genot_999_train["model"] = "UGENOT-K 0.999"
df_genot_999_train["data"] = "train"
df_genot_99_train = pd.read_csv("pancreas_unbalanced_99_genot_train.csv")
df_genot_99_train["model"] = "UGENOT-K 0.99"
df_genot_99_train["data"] = "train"
df_genot_95_train = pd.read_csv("pancreas_unbalanced_95_genot_train.csv")
df_genot_95_train["model"] = "UGENOT-K 0.95"
df_genot_95_train["data"] = "train"
df_genot_90_train = pd.read_csv("pancreas_unbalanced_90_genot_train.csv")
df_genot_90_train["model"] = "UGENOT-K 0.90"
df_genot_90_train["data"] = "train"
df_genot_80_train = pd.read_csv("pancreas_unbalanced_80_genot_train.csv")
df_genot_80_train["model"] = "UGENOT-K 0.80"
df_genot_80_train["data"] = "train"

df_genot_100_test = pd.read_csv("pancreas_unbalanced_100_genot_test.csv")
df_genot_100_test["model"] = "UGENOT-K 1.0"
df_genot_100_test["data"] = "test"
df_genot_999_test = pd.read_csv("pancreas_unbalanced_999_genot_test.csv")
df_genot_999_test["model"] = "UGENOT-K 0.999"
df_genot_999_test["data"] = "test"
df_genot_99_test = pd.read_csv("pancreas_unbalanced_99_genot_test.csv")
df_genot_99_test["model"] = "UGENOT-K 0.99"
df_genot_99_test["data"] = "test"
df_genot_95_test = pd.read_csv("pancreas_unbalanced_95_genot_test.csv")
df_genot_95_test["model"] = "UGENOT-K 0.95"
df_genot_95_test["data"] = "test"
df_genot_90_test = pd.read_csv("pancreas_unbalanced_90_genot_test.csv")
df_genot_90_test["model"] = "UGENOT-K 0.90"
df_genot_90_test["data"] = "test"
df_genot_80_test = pd.read_csv("pancreas_unbalanced_80_genot_test.csv")
df_genot_80_test["model"] = "UGENOT-K 0.80"
df_genot_80_test["data"] = "test"

df_genot_100_test_left = df_genot_100_test[df_genot_100_test["day"]==14.5]
df_genot_100_test_right = df_genot_100_test[df_genot_100_test["day"]==15.5]
df_genot_999_test_left = df_genot_999_test[df_genot_999_test["day"]==14.5]
df_genot_999_test_right = df_genot_999_test[df_genot_999_test["day"]==15.5]
df_genot_99_test_left = df_genot_99_test[df_genot_99_test["day"]==14.5]
df_genot_99_test_right = df_genot_99_test[df_genot_99_test["day"]==15.5]
df_genot_95_test_left = df_genot_95_test[df_genot_95_test["day"]==14.5]
df_genot_95_test_right = df_genot_95_test[df_genot_95_test["day"]==15.5]
df_genot_90_test_left = df_genot_90_test[df_genot_90_test["day"]==14.5]
df_genot_90_test_right = df_genot_90_test[df_genot_90_test["day"]==15.5]
df_genot_80_test_left = df_genot_80_test[df_genot_80_test["day"]==14.5]
df_genot_80_test_right = df_genot_80_test[df_genot_80_test["day"]==15.5]

df_genot_100_train_left = df_genot_100_train[df_genot_100_train["day"]==14.5]
df_genot_100_train_right = df_genot_100_train[df_genot_100_train["day"]==15.5]
df_genot_999_train_left = df_genot_999_train[df_genot_999_train["day"]==14.5]
df_genot_999_train_right = df_genot_999_train[df_genot_999_train["day"]==15.5]
df_genot_99_train_left = df_genot_99_train[df_genot_99_train["day"]==14.5]
df_genot_99_train_right = df_genot_99_train[df_genot_99_train["day"]==15.5]
df_genot_95_train_left = df_genot_95_train[df_genot_95_train["day"]==14.5]
df_genot_95_train_right = df_genot_95_train[df_genot_95_train["day"]==15.5]
df_genot_90_train_left = df_genot_90_train[df_genot_90_train["day"]==14.5]
df_genot_90_train_right = df_genot_90_train[df_genot_90_train["day"]==15.5]
df_genot_80_train_left = df_genot_80_train[df_genot_80_train["day"]==14.5]
df_genot_80_train_right = df_genot_80_train[df_genot_80_train["day"]==15.5]
```

```python
dfs_left = [df_genot_100_train_left,
            df_genot_999_train_left,
            df_genot_99_train_left,
            df_genot_95_train_left,
            df_genot_90_train_left,
            df_genot_80_train_left,
            df_genot_100_test_left,
            df_genot_999_test_left,
            df_genot_99_test_left,
            df_genot_95_test_left,
            df_genot_90_test_left,
            df_genot_80_test_left,
           ]
df = pd.concat(dfs_left)
```

```python
df1 = df[df["lineage"]=="A"]
df2 = df[df["lineage"]=="ED"]
```

```python
f, axes = plt.subplots(1, 2, figsize=(15, 4))
g = sns.barplot(df1, x="model", y="left_rescaling_0", hue="data", ax=axes[0])
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.set_ylabel("Reweighting of left marginals")
g.set_xlabel("Model")
g.set_title("Left rescaling function on train and test data on `A` lineage")
g = sns.barplot(df2, x="model", y="right_rescaling_0", hue="data", ax=axes[1])
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.set_ylabel("Reweighting of left marginals")
g.set_xlabel("Model")
g.set_title("Left rescaling function on train and test data on `ED` lineage")

#sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))
```
