```python
import pandas as pd
import seaborn as sns
```

```python
df_genot_100_train = pd.read_csv("pancreas_unbalanced_100_genot_train.csv")
df_genot_100_train["model"] = "UGENOT-FGW 1.0 train"
df_genot_999_train = pd.read_csv("pancreas_unbalanced_999_genot_train.csv")
df_genot_999_train["model"] = "UGENOT-FGW 0.999 train"
df_genot_99_train = pd.read_csv("pancreas_unbalanced_99_genot_train.csv")
df_genot_99_train["model"] = "UGENOT-FGW 0.99 train"
df_genot_95_train = pd.read_csv("pancreas_unbalanced_95_genot_train.csv")
df_genot_95_train["model"] = "UGENOT-FGW 0.95 train"
df_genot_90_train = pd.read_csv("pancreas_unbalanced_90_genot_train.csv")
df_genot_90_train["model"] = "UGENOT-FGW 0.90 train"
df_genot_80_train = pd.read_csv("pancreas_unbalanced_80_genot_train.csv")
df_genot_80_train["model"] = "UGENOT-FGW 0.80 train"

df_genot_100_test = pd.read_csv("pancreas_unbalanced_100_genot_test.csv")
df_genot_100_test["model"] = "UGENOT-FGW 1.0 test"
df_genot_999_test = pd.read_csv("pancreas_unbalanced_999_genot_test.csv")
df_genot_999_test["model"] = "UGENOT-FGW 0.999 test"
df_genot_99_test = pd.read_csv("pancreas_unbalanced_99_genot_test.csv")
df_genot_99_test["model"] = "UGENOT-FGW 0.99 test"
df_genot_95_test = pd.read_csv("pancreas_unbalanced_95_genot_test.csv")
df_genot_95_test["model"] = "UGENOT-FGW 0.95 test"
df_genot_90_test = pd.read_csv("pancreas_unbalanced_90_genot_test.csv")
df_genot_90_test["model"] = "UGENOT-FGW 0.90 test"
df_genot_80_test = pd.read_csv("pancreas_unbalanced_80_genot_test.csv")
df_genot_80_test["model"] = "UGENOT-FGW 0.80 test"

df_genot_100_test_left = df_genot_100_test[df_genot_100_test["day"]==14.5]
df_genot_100_test_right = df_genot_100_test[df_genot_100_test["day"]==15.5]
df_genot_999_test_left = df_genot_999_test[df_genot_999_test["day"]==14.5]
df_genot_999_test_right = df_genot_999_test[df_genot_999_test["day"]==15.5]
df_genot_99_test_left = df_genot_99_test[df_genot_99_test["day"]==14.5]
df_genot_99_test_right = df_genot_99_test[df_genot_99_test["day"]==15.5]
df_genot_95_test_left = df_genot_95_test[df_genot_95_test["day"]==14.5]
df_genot_95_test_right = df_genot_95_test[df_genot_95_test["day"]==15.5]
df_genot_90_test_left = df_genot_90_test[df_genot_90_test["day"]==14.5]
df_genot_90_test_right = df_genot_90_test[df_genot_90_test["day"]==15.5]
df_genot_80_test_left = df_genot_80_test[df_genot_80_test["day"]==14.5]
df_genot_80_test_right = df_genot_80_test[df_genot_80_test["day"]==15.5]

df_genot_100_train_left = df_genot_100_train[df_genot_100_train["day"]==14.5]
df_genot_100_train_right = df_genot_100_train[df_genot_100_train["day"]==15.5]
df_genot_999_train_left = df_genot_999_train[df_genot_999_train["day"]==14.5]
df_genot_999_train_right = df_genot_999_train[df_genot_999_train["day"]==15.5]
df_genot_99_train_left = df_genot_99_train[df_genot_99_train["day"]==14.5]
df_genot_99_train_right = df_genot_99_train[df_genot_99_train["day"]==15.5]
df_genot_95_train_left = df_genot_95_train[df_genot_95_train["day"]==14.5]
df_genot_95_train_right = df_genot_95_train[df_genot_95_train["day"]==15.5]
df_genot_90_train_left = df_genot_90_train[df_genot_90_train["day"]==14.5]
df_genot_90_train_right = df_genot_90_train[df_genot_90_train["day"]==15.5]
df_genot_80_train_left = df_genot_80_train[df_genot_80_train["day"]==14.5]
df_genot_80_train_right = df_genot_80_train[df_genot_80_train["day"]==15.5]

df_discrete_100_right = pd.read_csv("pancreas_unbalanced_100_discrete_right.csv")
df_discrete_100_right["right_rescaling_0"] = df_discrete_100_right["right_reweighting"]
df_discrete_100_right["model"] = "FGW discrete 1.0"
df_discrete_999_right = pd.read_csv("pancreas_unbalanced_999_discrete_right.csv")
df_discrete_999_right["right_rescaling_0"] = df_discrete_999_right["right_reweighting"]
df_discrete_999_right["model"] = "FGW discrete 0.999"
df_discrete_99_right = pd.read_csv("pancreas_unbalanced_99_discrete_right.csv")
df_discrete_99_right["right_rescaling_0"] = df_discrete_99_right["right_reweighting"]
df_discrete_99_right["model"] = "FGW discrete 0.99"
df_discrete_95_right = pd.read_csv("pancreas_unbalanced_95_discrete_right.csv")
df_discrete_95_right["right_rescaling_0"] = df_discrete_95_right["right_reweighting"]
df_discrete_95_right["model"] = "FGW discrete 0.95"
df_discrete_90_right = pd.read_csv("pancreas_unbalanced_90_discrete_right.csv")
df_discrete_90_right["right_rescaling_0"] = df_discrete_90_right["right_reweighting"]
df_discrete_90_right["model"] = "FGW discrete 0.90"
df_discrete_80_right = pd.read_csv("pancreas_unbalanced_80_discrete_right.csv")
df_discrete_80_right["right_rescaling_0"] = df_discrete_80_right["right_reweighting"]
df_discrete_80_right["model"] = "FGW discrete 0.80"

df_discrete_100_left = pd.read_csv("pancreas_unbalanced_100_discrete_left.csv")
df_discrete_100_left["model"] = "FGW discrete 1.0"
df_discrete_100_left["left_rescaling_0"] = df_discrete_100_left["left_reweighting"]
df_discrete_999_left = pd.read_csv("pancreas_unbalanced_999_discrete_left.csv")
df_discrete_999_left["model"] = "FGW discrete 0.999"
df_discrete_999_left["left_rescaling_0"] = df_discrete_999_left["left_reweighting"]
df_discrete_99_left = pd.read_csv("pancreas_unbalanced_99_discrete_left.csv")
df_discrete_99_left["model"] = "FGW discrete 0.99"
df_discrete_99_left["left_rescaling_0"] = df_discrete_99_left["left_reweighting"]
df_discrete_95_left = pd.read_csv("pancreas_unbalanced_95_discrete_left.csv")
df_discrete_95_left["model"] = "FGW discrete 0.95"
df_discrete_95_left["left_rescaling_0"] = df_discrete_95_left["left_reweighting"]
df_discrete_90_left = pd.read_csv("pancreas_unbalanced_90_discrete_left.csv")
df_discrete_90_left["model"] = "FGW discrete 0.90"
df_discrete_90_left["left_rescaling_0"] = df_discrete_90_left["left_reweighting"]
df_discrete_80_left = pd.read_csv("pancreas_unbalanced_80_discrete_left.csv")
df_discrete_80_left["model"] = "FGW discrete 0.80"
df_discrete_80_left["left_rescaling_0"] = df_discrete_80_left["left_reweighting"]
```

```python
dfs_left = [df_genot_100_train_left,
            df_genot_999_train_left,
            df_genot_99_train_left,
            df_genot_95_train_left,
            df_genot_90_train_left,
            df_genot_80_train_left,
            df_genot_100_test_left,
            df_genot_999_test_left,
            df_genot_99_test_left,
            df_genot_95_test_left,
            df_genot_90_test_left,
            df_genot_80_test_left,
            df_discrete_100_left,
            df_discrete_999_left,
            df_discrete_99_left,
            df_discrete_95_left,
            df_discrete_90_left,
            df_discrete_80_left,
           ]
df = pd.concat(dfs_left)
```

```python
g = sns.barplot(df, x="lineage", y="left_rescaling_0", hue="model")
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.set_ylabel("Reweighting of left marginals")
g.set_xlabel("Lineage")
sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))
```

```python
dfs_right = [df_genot_100_train_right,
            df_genot_999_train_right,
            df_genot_99_train_right,
            df_genot_95_train_right,
            df_genot_90_train_right,
            df_genot_80_train_right,
            df_genot_100_test_right,
            df_genot_999_test_right,
            df_genot_99_test_right,
            df_genot_95_test_right,
            df_genot_90_test_right,
            df_genot_80_test_right,
            df_discrete_100_right,
            df_discrete_999_right,
            df_discrete_99_right,
            df_discrete_95_right,
            df_discrete_90_right,
            df_discrete_80_right,
           ]
df2 = pd.concat(dfs_right)
```

```python
g = sns.barplot(df2, x="lineage", y="right_rescaling_0", hue="model")
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.set_ylabel("Reweighting of right marginals")
g.set_xlabel("Lineage")
sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))
```
