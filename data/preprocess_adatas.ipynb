```python
import scanpy as sc
```

```python
adata = sc.read("adata_pancreas_2019.h5ad")
```

```python
sc.pl.umap(adata, color=["day", "celltype"])
```

```python
adata.obs.celltype.cat.categories
```

```python
celltypes_to_remove_1 = ['Multipotent', 'Tip', 'Acinar', 'Trunk', 'Ductal']

adata_1 = adata[~adata.obs["celltype"].isin(celltypes_to_remove_1)]

adata_1.write("adata_pancreas_2019_endocrine.h5ad")
```

```python
celltypes_to_remove_2 = ['Multipotent', 'Tip', 'Acinar', 'Trunk']

adata_2 = adata[~adata.obs["celltype"].isin(celltypes_to_remove_2)]

adata_2.write("adata_pancreas_2019_endocrine_ductal.h5ad")
```
